# **Task 2**- View the distribution of the various features in the data set and calculate their central tendency

We will now look at the distributions of the various features in the data set

We wll also calculate appropriate measures of central tendency for these features

```python
#create a histogram of the "Fixed acidity" feature

plt.figure(figsize=(11,6))

sns.histplot(data = data , x ='fixed acidity',
             color ='orange',edgecolor='linen',alpha = 0.5,bins =5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel("Fixed Acidity")
plt.ylabel("Count")
plt.show()

```

## **Observations**
We observe that the histogram is normally distributed

Thee maximum count of values for fixed acidity lies in between 6 to 8.


lets see the measures of central tendency in working!



1.   Mean
2.   Median
3.  Mode



```python
# calculate the mean of "fixed acidity" feature
round(data['fixed acidity'].mean(),2)
```

```python
# calculate the median of "fixed acidity" feature
data['fixed acidity'].median()
```

```python
#create a histogram of the "Fixed acidity" feature and also show the mean and the median

plt.figure(figsize=(11,6))

sns.histplot(data = data , x ='fixed acidity',
             color ='orange',edgecolor='linen',alpha = 0.5,bins =5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel("Fixed Acidity")
plt.ylabel("Count")
plt.axvline(data['fixed acidity'].mean(),ymin=0,ymax=4000,color='blue',linestyle='dashed',linewidth=2,label='Mean')
plt.axvline(data['fixed acidity'].median(),ymin=0,ymax=4000,color='red',linestyle='dashed',linewidth=2,label='Median')
plt.legend()
plt.show()

```

**## Observations**

We can see that mean and median are clear representive of the data

We can choose either of  the parameters say mean as the measures of central tendency

We can choose either of the parameters say mean as the measure of central tendency

## Skewness
  Some distributions of data ,such as the bell curve or normal distribution,are symmetric. This means that the right and the left of the distributions are perfect mirror images of one another.Not every distribution of data is symmetric.Sets of data that are not symmetric are said to be asymmetric. The measures of how asymmetric a distribution can be is called skewness.

  The mean,median and mode are all measures of the center of a set of data. The skewness of the data can be determined by how these quantities are related to one another.



1.   If Skewness(S) = 0 ,then the distribution is normally distributed.
2.   If Skewness(S) > 0 ,then the distribution is positively skewed.
3.   If Skewness(S) < 0 ,then the distribution nagatively skewed.





```python
#create a distplot using "Volatile acidity" feature

plt.figure(figsize=(11,6))

sns.distplot(data['volatile acidity'],color ='blue')

plt.title("Distplot of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Count")
plt.show()
```

### Observations:
The above plot shows the normal distribution.

The normal distribution is described by the mean and the standard deviation.

The normal distribution is often refferred to as a "bell curve" because of its Shape:


*   The Median and mean are equal
*   It has only one mode
*   It is symmetric,meaning it decreases the same amount on the left and the right of the centre



We can calculate skewness too using skew () function in python

```python
#calculate skewness of "volatile acidity"
data['volatile acidity'].skew()
```

**Observation**

We clearly see that the skewness value is grater than 1,hence it is positively skewed
